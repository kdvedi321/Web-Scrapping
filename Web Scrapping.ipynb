```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "http://www.elections.in/delhi/polling-stations/rithala.html"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
type(soup)
```

```python
title = soup.title
title
```

```python
rows = soup.find_all('tr')
```

```python
list_all = []
for row in rows:
    row_td = row.find_all('td')
    str_cell = str(row_td)
    cleantext = BeautifulSoup(str_cell, "lxml").get_text()
    list_all.append(cleantext)
```

```python
a = pd.DataFrame(list_all)
```

```python
b = a[0].str.split(',')
b.head(10)
```

```python
b[0] = b[0].str.strip('[')
b.head(10)
```

```python
b.to_csv("WebScrap.csv", encoding='utf-8')
```
